```python
import pandas as pd
import numpy as np
import regex as re
```

```python
df=pd.read_excel('Data_Train.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Total_Stops']=df['Total_Stops'].apply(lambda x:re.sub(r'non-stop',r'0',str(x)))
df['Total_Stops']=df['Total_Stops'].apply(lambda x:re.sub(r'\D','',x))
df['Route'] = df['Route'].astype(str).apply(lambda x: re.sub(r'[^\p{Latin}\s]', '-', x))
```

```python
df.head()
```

```python
df.to_excel('Data_Train_Clean.xlsx',index=False)
```
